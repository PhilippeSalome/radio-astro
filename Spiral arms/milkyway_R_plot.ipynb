```python
from main_functions import *
from astropy.io import fits 
#warnings.filterwarnings('ignore')
%matplotlib inline

#
target_url = 'http://cdsarc.u-strasbg.fr/viz-bin/nph-Cat?-plus=-%2b&VIII/76/lab.fit'
hdulist = fits.open(target_url)

prihdr=hdulist[0].header
data2=hdulist[0].data
#
tetamin=-180. #borne inferieure de l'intervalle sur l'axe des x
tetamax=180. #borne superieure
tetaincr=1 #pas sur l'axe des x
zmin=-5. #borne inferieure de l'intervalle sur l'axe des y
zmax=5. #borne superieure
zincr=0.1 #pas sur l'axe des y
R=5. #hauteur du plan dans lequel on veut voir la galaxie
dV=1.
#
teta=arange(tetamin,tetamax,tetaincr)
z=arange(zmin,zmax,zincr)
TETA,Z = np.meshgrid(teta,z)
#
A=milkyway_R(R,tetamin,tetamax,tetaincr,zmin,zmax,zincr,dV,data2,prihdr) #appel de la fonction qui donne un plot du plan orthogonal a l'axe des z
#
plt.figure(1, figsize=(14,11))
plt.pcolormesh(TETA,Z,A,vmin=-4,vmax=-2, cmap='gist_ncar') # si on ne veut pas d'interpolation
#plt.figure(0)
#plt.imshow(A, interpolation='bicubic', origin='lower', extent=[zmin,zmax,tetamin,tetamax],vmin=-4,vmax=-2) # plot avec interpolation
plt.ylabel('Z [kpc]')
plt.xlabel('TETA [Degres]')
plt.title('Distribution en temperature de HI pour un rayon R='+str(R)+' fixe')
plt.colorbar()
plt.show()
```
